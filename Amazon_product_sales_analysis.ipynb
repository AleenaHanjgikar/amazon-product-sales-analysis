## ***PHASE 3 :—***
### ***TASK 4: Product Popularity Score***

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Normalize each component to 0–1 scale
df['norm_reviews'] = scaler.fit_transform(df[['number_of_reviews']])
df['norm_bought']  = scaler.fit_transform(df[['bought_in_last_month']])
df['norm_rating']  = scaler.fit_transform(df[['rating']])

# Weighted composite score (reviews=40%, bought=40%, rating=20%)
df['popularity_score'] = (
    0.40 * df['norm_reviews'] +
    0.40 * df['norm_bought']  +
    0.20 * df['norm_rating']
).round(4)

# Popularity Tier
pop_bins   = [0, 0.25, 0.50, 0.75, 1.01]
pop_labels = ['Low', 'Moderate', 'High', 'Viral']
df['popularity_tier'] = pd.cut(df['popularity_score'], bins=pop_bins, labels=pop_labels)

# Verify
print("Popularity Score Stats:")
print(df['popularity_score'].describe().round(4))
print()
print("Popularity Tier Counts:")
print(df['popularity_tier'].value_counts())
print()

# Top 10 Most Popular Products
top10 = df.nlargest(10, 'popularity_score')[['title', 'rating', 'bought_in_last_month', 'number_of_reviews', 'popularity_score']]
top10['title'] = top10['title'].str[:60] + '...'
print("Top 10 Most Popular Products:")
top10
```

#### Remove duplicates: 

```python
# Drop duplicate products (keep the one with highest popularity score)
df = df.sort_values('popularity_score', ascending=False).drop_duplicates(subset='title', keep='first')
df = df.reset_index(drop=True)

print(f"Dataset shape after deduplication: {df.shape}")
print(f"\nTop 5 after dedup:")
df.nlargest(5, 'popularity_score')[['title', 'popularity_score']].assign(
    title=lambda x: x['title'].str[:55]
)
```

## 7.1 Pricing Strategy
- The median price ($69.99) is far below the average ($148.74), confirming a heavily right-skewed market where a few premium products distort the average.
- Cheaper products (Very Low Price <$50) receive the highest average discounts (23.19%) while Premium products get the lowest (NaN — rarely discounted).
- Best Sellers average just $89.97 — affordability is the #1 driver of the Best Seller badge.

## 7.2 What Makes a Best Seller?
- Best Sellers have 6.7x more reviews (35,879 vs 5,296) — social proof is the strongest signal.
- Best Sellers sell 9.5x more units per month (9,589 vs 1,012) — volume compounds the badge.
- Best Sellers are rated higher (4.54 vs 4.44) — quality cannot be ignored.
- Best Sellers offer slightly higher discounts (25.18% vs 21.10%) — competitive pricing matters.
- Formula: High Quality + Low Price + High Volume + Strong Reviews = Best Seller


## 7.3 Sponsored vs Organic
- Only 3.8% of products are sponsored, yet they sell 74% more units (2,056 vs 1,183).
- Sponsored products are priced higher ($193 vs $147) — sellers invest in ads for premium items.
- Ratings are nearly identical (4.40 vs 4.45) — ads cannot compensate for poor quality.
- Sponsoring is a volume accelerator, not a quality substitute.

## 7.4 Discount Insights
- Only 33.3% of products carry a discount — majority are sold at full price.
- Medium discounts (10–30%) are the most common strategy (1,601 products).
- Discount percentage has almost zero correlation with sales volume (0.050) — discounts alone don't drive sales.
- The maximum discount of 99.96% is likely a data anomaly or an extreme flash sale.

## 7.5 Coupon Strategy
- Only 2% of products use coupons — extremely rare tactic.
- Couponed products actually sell LESS (1,059 vs 1,220 units) — coupons appear to be used by struggling products, not market leaders.
- Coupons are associated with lower-priced ($120 vs $149) and lower-rated (4.39 vs 4.45) products.

## 7.6 Popularity & Virality
- 98.6% of products sit in the Low popularity tier — the Amazon marketplace is extremely top-heavy.
- Only 2 products achieved Viral status, averaging 745,687 reviews and 100,000 units/month.
- Viral products are priced at just $14.74 on average — ultra-affordable everyday essentials dominate.
- Popularity correlates most strongly with units bought (0.807) and reviews (0.656) — volume is king.

## 7.7 Rating Insights
- Ratings are consistently high across the board (avg 4.45) — the Amazon marketplace self-selects quality products.
- Cheaper products rate slightly higher (Very Low: 4.51 vs High Price: 4.36) — budget buyers have lower expectations or everyday essentials simply deliver.
- Most common rating is 4.6 — the sweet spot for Amazon product success.
